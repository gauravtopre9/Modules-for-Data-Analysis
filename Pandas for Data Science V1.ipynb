### Summary Statistics

```python
#calculating Mean 1.0
df['Profit'].mean()
```

```python
#calculating Mean 2.0
df[['Profit','Quantity']].mean()
```

```python
#calculating Median 
df['Quantity'].median()
```

```python
#Statistical Summary

df[['Profit', 'Sales']].describe()
```

```python
df.agg({
    'Profit' : ['min','max','mean','median'],
    'Sales' : ['min','max','mean','median']
})
```

```python
df[df['Region']=='Central'].agg({
    'Profit' : ['min','max','mean','median'],
    'Sales' : ['min','max','mean','median']
})
```

**Applying Groupby**

```python
df[['Region','Profit']].groupby('Region').mean().sort_values(by = 'Profit')
```

```python
df.groupby(['Ship Mode','Region'])['Sales'].mean()
```

**Counting Number of Records by Category**

```python
#orders per Region
df['Region'].value_counts()
```

```python
#cities in each Region
df.groupby('Region')['City'].count()
```

### Reshaping The Layout of Table

```python
#sorting on the basis of city name

df.sort_values(by = 'City').tail()
```

```python
#sorting data on the basis of sales and profit

df.sort_values(by = ['Sales','Profit'], ascending = False).head()
```

```python
#pivot Tables

df_cen = df[df['Region']=='Central']
```

```python
df_cen.sort_index().groupby(['Ship Mode']).tail()
```
